## üìä –ö—Ä–æ–∫ 4: –ê–Ω–∞–ª—ñ–∑ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ñ–≤

```python
import json
import matplotlib.pyplot as plt

# –ó–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—è —ñ—Å—Ç–æ—Ä—ñ—ó –Ω–∞–≤—á–∞–Ω–Ω—è
with open('weights/advanced_history.json', 'r') as f:
    history = json.load(f)

# –ü—ñ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–∏—Ö
epochs = [h['epoch'] for h in history]
train_loss = [h['train']['loss'] for h in history]
val_loss = [h['val']['loss'] for h in history]
train_acc = [h['train']['cell_accuracy'] for h in history]
val_acc = [h['val']['cell_accuracy'] for h in history]
board_acc = [h['val']['board_accuracy'] for h in history]
lr = [h['lr'] for h in history]

# –í—ñ–∑—É–∞–ª—ñ–∑–∞—Ü—ñ—è
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Loss
axes[0, 0].plot(epochs, train_loss, label='Train Loss', marker='o')
axes[0, 0].plot(epochs, val_loss, label='Val Loss', marker='s')
axes[0, 0].set_xlabel('Epoch')
axes[0, 0].set_ylabel('Loss')
axes[0, 0].set_title('CNN Advanced - Loss')
axes[0, 0].legend()
axes[0, 0].grid(True)

# Cell Accuracy
axes[0, 1].plot(epochs, train_acc, label='Train Acc', marker='o')
axes[0, 1].plot(epochs, val_acc, label='Val Acc', marker='s')
axes[0, 1].set_xlabel('Epoch')
axes[0, 1].set_ylabel('Cell Accuracy')
axes[0, 1].set_title('CNN Advanced - Cell Accuracy')
axes[0, 1].legend()
axes[0, 1].grid(True)

# Board Accuracy
axes[1, 0].plot(epochs, board_acc, label='Board Acc', marker='D', color='green')
axes[1, 0].set_xlabel('Epoch')
axes[1, 0].set_ylabel('Board Accuracy')
axes[1, 0].set_title('CNN Advanced - Board Accuracy')
axes[1, 0].legend()
axes[1, 0].grid(True)

# Learning Rate
axes[1, 1].plot(epochs, lr, label='Learning Rate', marker='x', color='red')
axes[1, 1].set_xlabel('Epoch')
axes[1, 1].set_ylabel('Learning Rate')
axes[1, 1].set_title('CNN Advanced - Learning Rate Schedule')
axes[1, 1].legend()
axes[1, 1].grid(True)
axes[1, 1].set_yscale('log')

plt.tight_layout()
plt.savefig('weights/advanced_training_curves.png', dpi=300, bbox_inches='tight')
plt.show()

# –§—ñ–Ω–∞–ª—å–Ω—ñ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏
print("\n" + "="*60)
print("üìä –§–Ü–ù–ê–õ–¨–ù–Ü –†–ï–ó–£–õ–¨–¢–ê–¢–ò CNN ADVANCED")
print("="*60)
print(f"Best Validation Loss:       {min(val_loss):.4f}")
print(f"Best Cell Accuracy:         {max(val_acc):.2%}")
print(f"Best Board Accuracy:        {max(board_acc):.2%}")
print(f"Final Cell Accuracy:        {val_acc[-1]:.2%}")
print(f"Final Board Accuracy:       {board_acc[-1]:.2%}")
print(f"Final Learning Rate:        {lr[-1]:.6f}")
print("="*60)
```

## üß™ –ö—Ä–æ–∫ 5: –¢–µ—Å—Ç—É–≤–∞–Ω–Ω—è –º–æ–¥–µ–ª—ñ

```python
import torch
import numpy as np
import os
# –ü—Ä—è–º–∏–π —ñ–º–ø–æ—Ä—Ç - —É–Ω–∏–∫–∞—î–º–æ –∑–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—è torch_geometric
from models.cnn_advanced import CNNAdvanced

# ‚ö†Ô∏è –í–ê–ñ–õ–ò–í–û: –°–ø–æ—á–∞—Ç–∫—É –Ω–∞–≤—á—ñ—Ç—å –º–æ–¥–µ–ª—å (–∑–∞–ø—É—Å—Ç—ñ—Ç—å –ö—Ä–æ–∫ 3 –≤–∏—â–µ)!

# –ü–∞—Ä–∞–º–µ—Ç—Ä–∏ –º–æ–¥–µ–ª—ñ (–ø–æ–≤–∏–Ω–Ω—ñ –≤—ñ–¥–ø–æ–≤—ñ–¥–∞—é—Ç—å –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º –Ω–∞–≤—á–∞–Ω–Ω—è)
HIDDEN_CHANNELS = 128  # ‚Üê –í—ñ–¥–ø–æ–≤—ñ–¥–∞—î --hidden-channels 128
NUM_RESIDUAL_BLOCKS = 20  # ‚Üê –í—ñ–¥–ø–æ–≤—ñ–¥–∞—î --num-residual-blocks 20
DROPOUT = 0.25  # ‚Üê –í—ñ–¥–ø–æ–≤—ñ–¥–∞—î --dropout 0.25 (—Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü—ñ—è, –≤–∞–∂–ª–∏–≤–æ –¥–ª—è –≥–ª–∏–±–æ–∫–∏—Ö –º–µ—Ä–µ–∂)

# –ó–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—è –º–æ–¥–µ–ª—ñ
device = 'cuda' if torch.cuda.is_available() else 'cpu'

# –ü–µ—Ä–µ–≤—ñ—Ä–∫–∞ —ñ—Å–Ω—É–≤–∞–Ω–Ω—è —Ñ–∞–π–ª—É
checkpoint_path = 'weights/advanced_best.pth'
if not os.path.exists(checkpoint_path):
    raise FileNotFoundError(
        f"‚ùå –§–∞–π–ª {checkpoint_path} –Ω–µ –∑–Ω–∞–π–¥–µ–Ω–æ!\\n"
        "   –°–ø–æ—á–∞—Ç–∫—É –Ω–∞–≤—á—ñ—Ç—å –º–æ–¥–µ–ª—å, –∑–∞–ø—É—Å—Ç–∏–≤—à–∏ –ö—Ä–æ–∫ 3 (–∫–æ–º—ñ—Ä–∫—É –∑ !python train.py ...)"
    )

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –º–æ–¥–µ–ª—ñ
model = CNNAdvanced(hidden_channels=HIDDEN_CHANNELS, num_residual_blocks=NUM_RESIDUAL_BLOCKS, dropout=DROPOUT).to(device)

# –ó–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—è –≤–∞–≥
try:
    checkpoint = torch.load(checkpoint_path, map_location=device, weights_only=False)
    model.load_state_dict(checkpoint['model_state_dict'])
    model.eval()
    
    print(f"‚úÖ Model loaded from epoch {checkpoint['epoch']}")
    print(f"   Validation accuracy: {checkpoint['val_accuracy']:.2%}")
    print(f"   Hidden channels: {HIDDEN_CHANNELS}, Residual blocks: {NUM_RESIDUAL_BLOCKS}, Dropout: {DROPOUT}")
except RuntimeError as e:
    if 'size mismatch' in str(e):
        print("‚ùå –ü–û–ú–ò–õ–ö–ê: –†–æ–∑–º—ñ—Ä –º–æ–¥–µ–ª—ñ –Ω–µ –≤—ñ–¥–ø–æ–≤—ñ–¥–∞—î –∑–±–µ—Ä–µ–∂–µ–Ω–∏–º –≤–∞–≥–∞–º!")
        print(f"   –ü–æ—Ç–æ—á–Ω—ñ –ø–∞—Ä–∞–º–µ—Ç—Ä–∏: hidden_channels={HIDDEN_CHANNELS}, num_residual_blocks={NUM_RESIDUAL_BLOCKS}, dropout={DROPOUT}")
        print("   ")
        print("üí° –†–Ü–®–ï–ù–ù–Ø:")
        print("   1. –í–∏–¥–∞–ª—ñ—Ç—å —Å—Ç–∞—Ä—ñ –≤–∞–≥–∏: !rm -rf weights/advanced_*.pth")
        print("   2. –ù–∞–≤—á—ñ—Ç—å –º–æ–¥–µ–ª—å –∑–∞–Ω–æ–≤–æ –∑ –ø—Ä–∞–≤–∏–ª—å–Ω–∏–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏ (–ö—Ä–æ–∫ 3)")
        raise
    else:
        raise

# –¢–µ—Å—Ç–æ–≤–∏–π –ø–∞–∑–ª (—Å–∫–ª–∞–¥–Ω–∏–π)
test_puzzle = [
    [0, 0, 0, 2, 6, 0, 7, 0, 1],
    [6, 8, 0, 0, 7, 0, 0, 9, 0],
    [1, 9, 0, 0, 0, 4, 5, 0, 0],
    [8, 2, 0, 1, 0, 0, 0, 4, 0],
    [0, 0, 4, 6, 0, 2, 9, 0, 0],
    [0, 5, 0, 0, 0, 3, 0, 2, 8],
    [0, 0, 9, 3, 0, 0, 0, 7, 4],
    [0, 4, 0, 0, 5, 0, 0, 3, 6],
    [7, 0, 3, 0, 1, 8, 0, 0, 0]
]

# Inference
input_tensor = torch.tensor(test_puzzle, dtype=torch.long).unsqueeze(0).to(device)
with torch.no_grad():
    output = model(input_tensor)
    predictions = torch.argmax(output, dim=-1).squeeze().cpu().numpy() + 1
    
    # Confidence
    probs = torch.softmax(output, dim=-1)
    confidence = probs.max(dim=-1)[0].mean().item()

# –í–∏–≤–µ–¥–µ–Ω–Ω—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç—É
print("\nüéØ –¢–ï–°–¢ –Ü–ù–§–ï–†–ï–ù–°–£ (—Å–∫–ª–∞–¥–Ω–∏–π –ø–∞–∑–ª)")
print(f"Confidence: {confidence:.2%}\n")
print("Solved puzzle:")
for row in predictions:
    print(' '.join(map(str, row)))
```
